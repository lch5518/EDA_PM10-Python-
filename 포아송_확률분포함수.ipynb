## 확률분포함수 추정
data~f(x;seta)<br>
x_bar~N<br>
data를 가지고 모수를 어떻게 추정할 것인가.

왜도 : 치우침
왼쪽꼬리가 치우침(-)
오른쪽꼬리가 치우침(+)<br>

첨도 : 뽀족한 정도
좌우로 늘어짐(+)
위로 늘어짐(-)

```python
import scipy.stats as st
import numpy as np
from numpy import exp
from scipy.special import factorial
import matplotlib.pyplot as plt
%matplotlib inline

def mel_poi(mu,n):
    x = st.poisson.rvs(mu = mu, size = n)
    poisson_pmf = lambda y,mu:mu**y / factorial(y) * exp(-mu)
    y_values = range(0,mu*3)

    fig, ax = plt.subplots(figsize=(12,8))

    for mu in [mu-2,mu,mu+2]:
        distribution = []
        for y_i in y_values:
            distribution.append(poisson_pmf(y_i,mu))
        ax.plot(y_values,
               distribution,
               label = f'$\mu$={mu}',
               alpha = 0.5,
               marker = 'o',
               markersize=8)
    plt.hist(x,density=True,bins = 20)
    ax.grid()
    ax.set_xlabel('$y$',fontsize=14)
    ax.set_ylabel('$f(y\min/mu)$',fontsize=14)
    ax.axis(xmin=0,ymin=0)
    ax.legend(fontsize=14)
    plt.show()
```

```python
poisson_pmf = lambda y,mu:mu**y / factorial(y) * exp(-mu)
poisson_pmf(30,10)
```

```python
mel_poi(10,500)
```
