```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import font_manager,rc

font_name=font_manager.FontProperties(fname="C:/windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

mpl.rcParams['axes.unicode_minus'] = False

plt.rcParams["figure.figsize"] = (14,6) #set size of Graph
```

```python
mise10 = pd.read_csv('mise10.csv',encoding='ms949')
mise25 = pd.read_csv('mise25.csv',encoding='ms949')
```

```python
mise10 = mise10.rename(index=str, columns={"0.0": "보통", "1.0": "나쁨","2.0":"매우나쁨"})

mise25 = mise25.rename(index=str, columns={"0.0": "보통", "1.0": "나쁨","2.0":"매우나쁨"})
```

```python
mise10
```

```python
plt.title('16~18년도 광역시별 미세먼지(PM10) 보통인비율(%)')
ax = sns.boxplot(x="시도", y="보통", hue="연도",data=mise10)
ax.legend(bbox_to_anchor=(1.09,1.02))
```

```python
plt.title('16~18년도 광역시별 초미세먼지(PM2.5) 보통인비율(%)')
ax = sns.boxplot(x="시도", y="보통", hue="연도",data=mise25)
ax.legend(bbox_to_anchor=(1.09,1.02))
```
