```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import font_manager,rc

font_name=font_manager.FontProperties(fname="C:/windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

mpl.rcParams['axes.unicode_minus'] = False

plt.rcParams["figure.figsize"] = (10,6) #set size of Graph #(14,6)
```

```python
mise_fin = pd.read_csv('mise_fin.csv',encoding='ms949')
```

```python
mise_fin.head()
```

```python
result = pd.crosstab([mise_fin['연도'],mise_fin['시도'],mise_fin['월'],mise_fin['주말여부']],mise_fin['PM10group'],normalize='index')
```

```python
result2 = pd.crosstab([mise_fin['연도'],mise_fin['시도'],mise_fin['월'],mise_fin['주말여부']],mise_fin['PM25group'],normalize='index')
```

```python
result.head()
```

```python
result0 = result.reset_index()
```

```python
result22 = result2.reset_index()
```

```python
result0.head()
```

```python
plt.title('16~18년도 미세먼지(PM10) 나쁨기준')
ax = sns.boxplot(x="월", y=1.0, hue="연도", data=result0)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```

```python
plt.title('16~18년도 미세먼지(PM10) 매우나쁨기준')
ax = sns.boxplot(x="월", y=2.0, hue="연도", data=result0)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```

```python
plt.title('16~18년도 초미세먼지(PM25) 나쁨기준')
ax = sns.boxplot(x="월", y=1.0, hue="연도", data=result22)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```

```python
plt.title('16~18년도 초미세먼지(PM25) 매우나쁨기준')
ax = sns.boxplot(x="월", y=2.0, hue="연도", data=result22)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```
