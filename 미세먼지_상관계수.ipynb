```python
mise = pd.read_csv('mise0.csv',encoding = 'ms949')
```

```python
date = mise['측정일시'].astype(str)
mise['연도'] = date.str.slice(start=0, stop=4).astype(str)
mise['월'] = date.str.slice(start=4, stop=6).astype(str)
mise['일'] = date.str.slice(start=6, stop=8).astype(str)
mise['시간'] = date.str.slice(start=8, stop=10).astype(str)
```

```python
new = mise['지역'].str.split(' ', n = 1, expand = True) 
mise['시도'] = new[0]
```

```python
mise.head()
```

```python
mise.info()
```

```python
mise1 = mise[mise['시도'].values != '강원']
mise1 = mise1[mise1['시도'].values != '경기']
mise1 = mise1[mise1['시도'].values != '경남']
mise1 = mise1[mise1['시도'].values != '경북']
mise1 = mise1[mise1['시도'].values != '전북']
mise1 = mise1[mise1['시도'].values != '전남']
mise1 = mise1[mise1['시도'].values != '충남']
mise1 = mise1[mise1['시도'].values != '충북']
mise1 = mise1[mise1['시도'].values != '제주']
mise1 = mise1[mise1['시도'].values != '세종']
```

```python
mise1.info()
```

```python
mise_fin = mise1[mise1['시간']=='11']
```

```python
mise_fin=mise_fin.groupby(by=['시도','측정일시']).max()
mise_fin.reset_index(inplace=True)
```

```python
mise_fin.head()
```

```python
mise_fin.shape
```

```python
mise_fin.info()
```

```python
mise_fin['시도'].value_counts()
```

```python
mise_fin.loc[mise_fin['PM10']<=80,'PM10group'] = 0
mise_fin.loc[mise_fin['PM10']>80,'PM10group'] = 1
mise_fin.loc[mise_fin['PM10']>150,'PM10group'] = 2
```

```python
mise_fin.loc[mise_fin['PM25']<=35,'PM25group'] = 0
mise_fin.loc[mise_fin['PM25']>35,'PM25group'] = 1
mise_fin.loc[mise_fin['PM25']>75,'PM25group'] = 2
```

```python
mise_fin.head()
```

```python
mise_fin['PM10group'].value_counts()
```

```python
mise_fin.head()
```

```python
date = mise_fin['측정일시'].astype(str)
mise_fin['날짜'] = date.str.slice(start=0, stop=8).astype(str)

mise_fin['날짜2'] = pd.to_datetime(mise_fin['날짜'])

mise_fin['요일'] = mise_fin['날짜2'].dt.weekday

mise_fin['주말여부'] = np.where(mise_fin['요일']>=5, '주말','평일')
```

```python
mise_fin
```

```python
mise10 = pd.crosstab([mise_fin['시도'],mise_fin['연도'],mise_fin['주말여부']],mise_fin['PM10group'],normalize='index')

mise25 = pd.crosstab([mise_fin['시도'],mise_fin['연도'],mise_fin['주말여부']],mise_fin['PM25group'],normalize='index')
```

```python
mise10.to_csv('mise10.csv',encoding='ms949')
```

```python
mise25.to_csv('mise25.csv',encoding='ms949')
```
