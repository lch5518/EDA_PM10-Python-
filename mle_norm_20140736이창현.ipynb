```python
from scipy.stats import norm
import numpy as np

def mle_norm(mu,s,n):
    fig, ax = plt.subplots(figsize=(12,8))
    y=norm.rvs(size=n)
    x = np.linspace(-s, s, 100)
    ax.set_title('Population PDF & Histigram of Generating data')
    for mu in [mu-1,mu,mu+1]:
            ax.plot(x, norm(loc=mu).pdf(x),  lw=2, alpha=0.6, label=f'norm pdf($\mu$={mu})')
    ax.hist(y, density=True, histtype='stepfilled', alpha=0.2)
    ax.legend(loc='best', frameon=False, fontsize=12)
    plt.show()
```

```python
mle_norm(0,3,1000)
```
