```python
mise = pd.read_csv('mise161718.csv',encoding = 'ms949')

mise['날짜'] = mise['날짜'].astype(str)

date = mise['측정일시'].astype(str)
mise['날짜'] = date.str.slice(start=0, stop=8).astype(str)

mise['날짜2'] = pd.to_datetime(mise['날짜'])

mise['연도2'] = mise['날짜2'].dt.year
mise['월2'] = mise['날짜2'].dt.month
mise['요일2'] = mise['날짜2'].dt.weekday
```

```python
plt.title('16~18년도 광역시별 미세먼지(PM10)')
ax = sns.boxplot(x="시도", y="PM10", hue="연도2",
                 data=mise, order=['서울', '인천', '대전', '광주', '대구','울산','부산'])
ax.set_ylim(0,200)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.axhline(151,color='red',linestyle=':')
plt.axhline(81,color='red',linestyle=':')
```

```python
plt.title('16~18년도 광역시별 초미세먼지(PM2.5)')
ax = sns.boxplot(x="시도", y="PM25", hue="연도2",
                 data=mise, order=['서울', '인천', '대전', '광주', '대구','울산','부산'])
ax.set_ylim(0,150)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.axhline(76,color='red',linestyle=':')
plt.axhline(36,color='red',linestyle=':')
```

```python
plt.title('16~18년도 요일별 미세먼지(PM10)')
ax = sns.boxplot(x="요일2", y="PM10", hue="연도2",
                 data=mise)
ax.set_ylim(0,200)
ax.legend(bbox_to_anchor=(1.1,1.05))
plt.axhline(151,color='red',linestyle=':')
plt.axhline(81,color='red',linestyle=':')
```
