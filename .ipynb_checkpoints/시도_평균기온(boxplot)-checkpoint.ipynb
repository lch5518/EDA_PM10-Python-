```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import font_manager,rc

font_name=font_manager.FontProperties(fname="C:/windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

mpl.rcParams['axes.unicode_minus'] = False

plt.rcParams["figure.figsize"] = (14,6) #set size of Graph

ds=pd.read_csv('add.csv',names = ["0", "지점", "2"],encoding='ms949')

new = ds['2'].str.split(' ', n = 1, expand = True) 
ds['시도'] = new[0]

ds1=pd.read_csv('2018climate.csv',encoding='ms949')


ds0 = ds[['지점','시도']]
ds11 = ds1[['지점','평균기온(°C)']]
    
df = pd.merge(ds0,ds11)

sns.boxplot(x='시도',y='평균기온(°C)',data=df,palette="Set3")
plt.title('20140736이창현')
```
